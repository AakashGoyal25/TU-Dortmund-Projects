Create scatter plots of the features 'Dribbling' and 'Ball control', then apply z-normalization to both features, and finally applying PCA to the normalized 2-d data (keep both dimensions after the transformation).

```python
#TODO: make scatter plots of the original normalized and the transformed data


# Original Plot
plt.figure(0)
plt.title("Original plot")
plt.scatter( df['dribbling'],df['skill_ball_control'])

# Scaling the variables
df['dribbling'] = scale(df['dribbling'])
df['skill_ball_control'] = scale(df['skill_ball_control'])



# Plot after scaling the values
plt.figure(1)
plt.title("Scaled data plot")
plt.scatter( df['dribbling'],df['skill_ball_control'])

new_df = df[['dribbling','skill_ball_control' ]].dropna()
pca = PCA()
pca.fit(X =new_df)
cc = pd.DataFrame(pca.transform(X=new_df))


# Plot after z-transformation
plt.figure(3)
plt.title("Plot after PCA")
plt.scatter(cc[0],cc[1])
```

```python
# TODO: output the amount of explained variance of each component and remining variance after dimensionality reduction
print("Varience explained by each component",pca.explained_variance_ratio_)
```
